# DataFrame demo

This notebook demonstrates the Pandas DataFrame generated by a HoboReader instance.

## Create HoboReader instance and read in csv file

```python
from hoboreader import HoboReader
h=HoboReader('sample_hobo_data.csv')
```

## Create Pandas DataFrame

```python
df=h.get_dataframe()
df.head()
```

When the DataFrame is created, the following actions occur:

- the data from the csv file is placed in the columns of the dataframe
- any blank cells are converted to 'NaN'
- any numeric values are converted to numbers
- an index of datetimes is created from the 'Date Time' column and includes the timezone information 
- a multi-level column index is set up with the information from the header row in the csv file

## Quick plot of the temperature data
